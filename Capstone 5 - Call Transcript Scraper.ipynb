```python
import urllib.request, re
from bs4 import BeautifulSoup

url_re = re.compile(r'http\S+')
call_links = []

# Grab XML sitemap data to get links for each article
for year in [2017, 2018, 2019, 2020]:
    for version in range(1,21):
        if version < 10:
            version = "0%.0f" % version
        url = 'https://www.fool.com/sitemap/{}/{}'.format(year, version)
        response = str(urllib.request.urlopen(url).read())
        
        # Use Regex to get all link objects from text data
        urls = url_re.findall(response)
        
        # Strip extraneous data and append to list of links to crawl
        for link in urls:
            link = re.sub(r'<\S+>', '', link)
            link = re.sub(r'\\n', '', link)
            if 'call-transcripts' in link and link not in call_links:
                call_links.append(link)
```

```python
import urllib.request, requests
from bs4 import BeautifulSoup

print(len(call_links))
files_written = 0
bad_link = 0

for link in call_links:
    filename = link[58:-5]
    filepath = "/Users/jkovach/Downloads/earnings-call-transcripts/" + filename + ".txt"
    try:
        page = requests.get(link)
        soup = BeautifulSoup(page.content, 'html.parser')
        g = soup.find("span", class_="article-content").get_text()
        with open(filepath, 'w') as file:
            file.write(g)
        files_written += 1
        if files_written % 100 == 0:
            print("Files Written: {}".format(files_written))
    except:
        bad_link +=1
        print("Bad Link; Total: {}".format(bad_link))
        continue

print("complete")
```
